```python
dataset_Train=pd.read_csv('/content/drive/MyDrive/HAR/train.csv')
dataset_Test=pd.read_csv('/content/drive/MyDrive/HAR/test.csv')
dataset_Train=dataset_Train.drop(['subject'],axis=1)
dataset_Test=dataset_Test.drop(['subject'],axis=1)
features_used=dataset_Train.columns[:-1]
print(features_used)
X_trainB=dataset_Train.iloc[:,:-1].values
Y_trainB=dataset_Train.iloc[:,-1].values
X_testB=dataset_Test.iloc[:,:-1].values
Y_testB=dataset_Test.iloc[:,-1].values
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_trainB=sc.fit_transform(X_trainB)
X_testB=sc.transform(X_testB)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y_trainB=le.fit_transform(Y_trainB)
Y_testB=le.transform(Y_testB)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from niapy.problems import Problem
from niapy.task import Task
from niapy.algorithms.basic import ArtificialBeeColonyAlgorithm
from sklearn.naive_bayes import GaussianNB

class FeatureSelection(Problem):
    def __init__(self, X_train, y_train, alpha=0.99):
        super().__init__(dimension=X_train.shape[1], lower=0, upper=1)
        self.X_train = X_train
        self.y_train = y_train
        self.alpha = alpha

    def _evaluate(self, x):
        selected = x > 0.6
        num_selected = selected.sum()
        if num_selected == 0:
            return 1.0
        accuracy = cross_val_score(GaussianNB(), self.X_train[:, selected], self.y_train, cv=2, n_jobs=-1).mean()
        score = 1 - accuracy
        num_features = self.X_train.shape[1]
        return self.alpha * score + (1 - self.alpha) * (num_selected / num_features)
problem = FeatureSelection(X_trainB, Y_trainB,alpha=0.99)
task = Task(problem, max_iters=100)
algorithm = ArtificialBeeColonyAlgorithm(population_size=10, seed=1234)
best_features, best_fitness = algorithm.run(task)

selected_features = best_features > 0.6
print('Number of selected features:', selected_features.sum())
filtered = np.array(features_used)[np.array(selected_features)]
print('Selected Features are:')
print(filtered)

Selected_AC_X_Train=dataset_Train[filtered]
Selected_AC_Y_Train=dataset_Train['Activity']

Selected_AC_X_Test=dataset_Test[filtered]
Selected_AC_Y_Test=dataset_Test['Activity']

################## Feature Selection of DE ############
X_train=Selected_AC_X_Train.iloc[:,:].values
Y_train=Selected_AC_Y_Train.values
X_test=Selected_AC_X_Test.iloc[:,:].values
Y_test=Selected_AC_Y_Test.values

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y_train=le.fit_transform(Y_train)
Y_test=le.transform(Y_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score,confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
import seaborn as sns
svm_model= SVC(C=0.01,kernel='rbf')
nb_model=GaussianNB()

svm_model.fit(X_trainB,Y_trainB)
nb_model.fit(X_trainB,Y_trainB)

y_predSVM=svm_model.predict(X_testB)
y_prednb=nb_model.predict(X_testB)


print('Model Performance for Before Optimization for SVM')
print('Accuracy:', accuracy_score(Y_testB, y_predSVM))
print('F1 score:', f1_score(Y_testB, y_predSVM,average='weighted'))
print('Recall:', recall_score(Y_testB, y_predSVM,average='weighted'))
print('Precision:', precision_score(Y_testB, y_predSVM,average='weighted'))
data = confusion_matrix(le.inverse_transform(Y_testB), le.inverse_transform(y_predSVM))
df_cm = pd.DataFrame(data, columns=np.unique(le.inverse_transform(Y_testB)), index = np.unique(le.inverse_transform(y_predSVM)))
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'
plt.figure(figsize = (5,5))
sns.set(font_scale=0.7)#for label size
sns.heatmap(df_cm, cmap="Blues", annot=True,annot_kws={"size": 8})# font size
plt.title('Confusion Matrix')
print('End of Results')



print('Model Performance for Before Optimization for NB')
print('Accuracy:', accuracy_score(Y_testB, y_prednb))
print('F1 score:', f1_score(Y_testB, y_prednb,average='weighted'))
print('Recall:', recall_score(Y_testB, y_prednb,average='weighted'))
print('Precision:', precision_score(Y_testB, y_prednb,average='weighted'))
data = confusion_matrix(le.inverse_transform(Y_testB), le.inverse_transform(y_prednb))
df_cm = pd.DataFrame(data, columns=np.unique(le.inverse_transform(Y_testB)), index = np.unique(le.inverse_transform(y_prednb)))
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'
plt.figure(figsize = (5,5))
sns.set(font_scale=0.7)#for label size
sns.heatmap(df_cm, cmap="Blues", annot=True,annot_kws={"size": 8})# font size
plt.title('Confusion Matrix')
print('End of Results')
```
